This program takes the Uber data and rounds the time data from minutes into hours in order to match the format of the weather data

```python
import pandas as pd
import numpy as np
import collections
import matplotlib.pyplot as plt

from datetime import date


temperature_to_load = "temperature.csv"

temperature_data = pd.read_csv(temperature_to_load)
temperature_data.head(5)
```

```python
# File to Load (Remember to Change These)
uber_to_load = "ubernyc.csv"

uber_data = pd.read_csv(uber_to_load)
uber_data.head(5)
```

```python
#from ubernyc.csv merge the columns DATE and TIME and round the time hours
uber_data.index = pd.to_datetime(uber_data["DATE"] + " " + uber_data["TIME"])
uber_data.head(5)
```

```python
ut = uber_data.index.round("H")
ut = pd.DataFrame(ut)
#print(ut)
```

Export the ut into a csv
with open('test1.csv', mode='w') as output_data_file:
ut.to_csv(output_data_file, index_label="Index")

```python
#take column from ut df with timestamps and make it into a list.
list1 = []
list1 = ut[0].tolist()
```

```python
#count the number of uber rides per hour. The output looks like a library key: value, Key is the time and value the number of rides during that time
counter=collections.Counter(list1)

df_time_as_header = pd.DataFrame(counter, index=[0])
#transponde df_tdf_time_as_headerime_as_header to get correct format, one hour per row insated of one hour per column
df_final = df_time_as_header.T
df_final.head(5)
```

```python
#rename axis so that later the two df (Uber and Temperature) can  be merged using their index
df_final.rename_axis('datetime')

#rename column, by default 0, to 'number_of_rides' and change the name of the df to be more descriptive
df_number_of_rides_per_hour = df_final.rename(columns={0: 'number_of_rides'})
```

Converting K to F

```python
def kelvin_to_fahr(temp_kelvin):
    temp_fahr = (((temp_kelvin - 273) * 9/5) + 32)
    return temp_fahr
```

```python
#select columns 'datetime' and 'New York' from tempreature data and store in a new df NYC_df
NYC_df = temperature_data[['datetime', 'New York']].copy()

#make datetime the index for 'NYC_df' df
NYC_df.set_index('datetime', inplace=True)

#rename New York column to be Temperature in K
NYC_df = NYC_df.rename(columns={'New York': 'Temperature K'})

NYC_df = NYC_df.join(df_number_of_rides_per_hour, how='inner')


NYC_df["Temperature K"] = kelvin_to_fahr(NYC_df["Temperature K"])
NYC_df = NYC_df.rename(columns={'Temperature K': 'Temperature F', 'number_of_rides': 'Number of Rides'})
NYC_df.head()
```

```python
w_description_to_load = "weather_description.csv"

w_description = pd.read_csv(w_description_to_load)
w_description.head(5)
```

```python
NYCw_df = w_description[['datetime', 'New York']]

NYCw_df.set_index('datetime', inplace=True)

NYCw_df = NYCw_df.rename(columns={'New York': 'Weather Description'})

NYC_df_joined = NYC_df.join(NYCw_df, how='inner')

NYC_df_joined.head(5)
```

Creating index for date/time

```python
#NYC_df_joined.index
NYC_df_joined['index1'] = NYC_df_joined.index

NYC_df_joined['index1'] = NYC_df_joined['index1'].astype(str)
#NYC_df_joined.index1 = NYC_df_joined.index1.str[:19]
NYC_df_joined.head(5)
```

Identify peak time

```python
peak_time_filter_list = ['01:00:00','02:00:00', '03:00:00', '04:00:00', '07:00:00', '08:00:00', '09:00:00', '12:00:00', '16:00:00', '17:00:00', '18:00:00']
#filter_list = ['08:00:00']
peak_rows_df = pd.DataFrame()
for item in peak_time_filter_list:
    temp_peak_df = NYC_df_joined[NYC_df_joined.index1.str.contains(item)]
    peak_rows_df = peak_rows_df.append(temp_peak_df)
peak_rows_df = peak_rows_df.sort_index()
#peak_rows_df.to_csv("DF.csv", sep=',')
peak_rows_df.head(5)
```

```python
peak_rows_ew_df = peak_rows_df[peak_rows_df.index.dayofweek < 7]
peak_rows_ew_df['index1'] = pd.to_datetime(peak_rows_ew_df['index1'])
pd.options.mode.chained_assignment = None
peak_rows_ew_df
peak_rows_ew_df['weekday'] = peak_rows_ew_df['index1'].dt.weekday_name
peak_rows_ew_df['Weather Description'].unique()
peak_rows_ew_df.to_csv("DF1.csv", sep=',')
```

Adding day of a week, filter on day and hour (this should change after screenshot is taken, or we can repeat the code 15x)

```python
peak_with_filters = peak_rows_ew_df.loc[peak_rows_ew_df.weekday == 'Sunday']
peak_with_filters['index1'] = peak_with_filters['index1'].astype(str)
peak_with_filters = peak_with_filters[peak_with_filters.index1.str.contains("04:00:00")]
peak_with_filters['Weather Description'].unique()
```

Generating plot for specific hour in a specific day for every week of our data set. Colors by weather. 

```python
x_axis = np.arange(peak_with_filters['index1'].count())
#print(x_axis)
tick_locations = [value for value in x_axis]
xdates = peak_with_filters.index1.str[:10].tolist()
plt.xticks(tick_locations, xdates, rotation='vertical')
barplt = plt.bar(x_axis, peak_with_filters['Number of Rides'], alpha=0.5, align="center")
#plt.show()
weatherlist = peak_with_filters['Weather Description'].unique().tolist()
#print(weatherlist)
barcolors = ['red', 'blue', 'green', 'yellow', 'orange', 'pink', 'grey', 'maroon', 'brown', 'navy', 'lavender', 'lime', 'purple', 'olive', 'apricot', 'teal']
weather_color_dict = {}
i = 0
for item in range(0, len(weatherlist)):
    weather_color_dict.update({weatherlist[item]: barcolors[i]})
    i += 1
#print(weather_color_dict)
j = 0
for item in xdates:
    temp_date_df = peak_with_filters[peak_with_filters.index1.str.contains(item)]
    bar_color = weather_color_dict.get(temp_date_df['Weather Description'].tolist()[0])
    #print(bar_color)
    barplt[j].set_color(bar_color)
    j += 1

import matplotlib.patches as mpatches

patchList = []
for key in weather_color_dict:
        data_key = mpatches.Patch(color=weather_color_dict[key], label=key)
        patchList.append(data_key)

plt.legend(handles=patchList, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.tight_layout()
plt.savefig("analysis/Sunday4.jpg")
plt.show()
```
